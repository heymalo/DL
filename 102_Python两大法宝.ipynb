# 1. Python两大法宝

① Python3.6.3相当于一个package，package里面有不同的区域，不同的区域有不同的工具。

② Python语法有两大法宝：dir()、help() 函数。
 - dir()：打开，看见里面有多少分区、多少工具。
 - help()：说明书。

```python
import torch
print(torch.cuda.is_available())
help(torch.cuda.is_available) # 查看 torch.cuda.is_available 的用法
dir(torch.cuda)  # 查看torch包中有哪些区、有哪些工具import torch

# torch 是包/模块，torch.cuda 是包/模块，torch.cuda.is_available 是函数/方法
# 模块下面有类/函数，区分方法调用type(x),看输出类型
```
